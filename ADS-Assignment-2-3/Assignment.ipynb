```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read dataframe
employee= pd.read_csv('./WA_Fn-UseC_-HR-Employee-Attrition.csv')

employee
```

```python
#calculation monthly income

# daily rate *5days a week* 4 weeks a month
income = (employee.DailyRate*5)*4
income

employee['income']=income

employee
```

```python
#select Dist Department 
df = employee[['DistanceFromHome','Department','Attrition']]


df
```

```python
#unique job roles
jobrole = pd.unique(employee['Department'])
jobrole
```

```python
#distance by jobrole ,distance from home and attrition
df1 =pd.DataFrame()

df1=pd.DataFrame(df.groupby(['DistanceFromHome','Department'])['Attrition'].value_counts().reset_index(name='a'))

df1



df1

df4=df1[df1['Attrition'].loc[:137]=="Yes"]

df4
```

```python
#visualization of distance from home by jobrole and attrition 



sns.set_style('whitegrid')

fig=sns.catplot( x='DistanceFromHome', y='a',hue='Department',kind='bar' ,data=df4,height=7,aspect=2)

plt.xlabel('Distance From Home')
plt.ylabel('Attrition')
plt.title('distance from home by jobrole and attrition ')
```

```python
#monthly income by education and attrition

df2 = employee[['Education','Attrition','income']]

df2
```

```python
#breakdown of monthly income by education level and attrition
df3 = df2.groupby(['Education','Attrition'])['income'].mean().reset_index()

df3
```

```python
#visualization of mean monthly income by Education and Attrition
sns.set_style('darkgrid')

sns.catplot( y="income" ,x='Education' ,hue="Attrition", kind="bar",  data= df3, aspect=2)

plt.xlabel('Education')
plt.ylabel('Income')
plt.title("mean monthly income by Education and Attrition")
```
